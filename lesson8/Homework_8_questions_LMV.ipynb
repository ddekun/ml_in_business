#### 1. Если мы рассчитали 95% доверительный интервал для среднего значения, то какие из следующих утверждений являются верными?

Выберите несколько вариантов и попытайтесь объяснить свой выбор:

1) Если многократно повторять эксперимент, то 95 % выборочных средних значений будут принадлежать рассчитанному нами доверительному интервалу.

2) Мы можем быть на 95% уверены, что среднее значение в генеральной совокупности принадлежит рассчитанному доверительному интервалу.

3) Если многократно повторять эксперимент, для каждой выборки рассчитывать свой доверительный интервал, то в 95 % случаев истинное среднее будет находиться внутри доверительного интервала.

4) Среднее значение в генеральной совокупности точно превышает нижнюю границу 95% доверительного интервала.

5) Среднее значение в генеральной совокупности точно принадлежит рассчитанному доверительному интервалу.

#### Ответ:
1. Верно. Доверительный интервал позволяет оценить диапазон значений, в котором с 
определенной вероятностью находится истинное значение параметра. Если эксперимент будет многократно повторен, то 95% выборочных средних значений будут принадлежать рассчитанному нами доверительному интервалу.

2. Неверно. Доверительный интервал представляет собой диапазон значений, в котором находится истинное значение параметра с определенной вероятностью (обычно 95%). Нельзя с уверенностью утверждать, что среднее значение в генеральной совокупности находится в рассчитанном доверительном интервале с вероятностью 95%.

3. Верно. Если многократно повторять эксперимент и для каждой выборки рассчитывать свой доверительный интервал, то в 95% случаев истинное среднее будет находиться внутри доверительного интервала. Это связано с тем, что при повторении эксперимента мы получаем выборки из той же генеральной совокупности, и поэтому доверительный интервал будет содержать истинное среднее в большинстве случаев.

4. Неверно. Доверительный интервал представляет собой диапазон значений, в котором находится истинное значение параметра с определенной вероятностью (обычно 95%). Нижняя граница доверительного интервала не является точной оценкой истинного среднего значения в генеральной совокупности, а лишь одной из возможных оценок.

5. Неверно. Доверительный интервал представляет собой диапазон значений, в котором находится истинное значение параметра с определенной вероятностью (обычно 95%). Среднее значение в генеральной совокупности может находиться как внутри доверительного интервала, так и за его пределами.
